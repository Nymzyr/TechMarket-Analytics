## Load Source Data

All datasets are stored in the `06_Datasets/raw` directory.

The following step loads every available CSV file into memory.

```python
DATA_PATH = Path("../06_Datasets/raw")
```

```python
dataset_files = {
    "customers": "olist_customers_dataset.csv",
    "orders": "olist_orders_dataset.csv",
    "order_items": "olist_order_items_dataset.csv",
    "order_payments": "olist_order_payments_dataset.csv",
    "order_reviews": "olist_order_reviews_dataset.csv",
    "products": "olist_products_dataset.csv",
    "sellers": "olist_sellers_dataset.csv",
    "geolocation": "olist_geolocation_dataset.csv",
    "categories": "product_category_name_translation.csv"
}
```

```python
datasets = {}
for name, filename in dataset_files.items():
    datasets[name] = pd.read_csv(DATA_PATH / filename)
```

## Dataset Overview

This section summarizes the dimensions of each dataset before detailed inspection.

```python
overview = []

for name, df in datasets.items():
    overview.append({
        "Dataset": name,
        "Rows": df.shape[0],
        "Cols": df.shape[1],
        "Memory (mb)": round(df.memory_usage(deep=True).sum()/1024**2, 2)
    })

overview = pd.DataFrame(overview)
overview
```

The project consists of nine relational datasets.

The largest table is Order Items.

The smallest one is Category Translation.

The datasets represent different entities of an e-commerce platform.

## Data Structure Review

Each dataset is briefly inspected to understand its schema and business purpose.

```python
for name, df in datasets.items():
    print("=" * 80)
    print(name.upper())
    print("=" * 80)
    display(df.head(3))
    display(df.dtypes.to_frame("Data Type"))
    print(f"Rows: {df.shape[0]:,}")
    print(f"Columns: {df.shape[1]}")
    print(df.info())
    print("\n")
```

## Data Quality Assessment

The following checks evaluate the overall quality of the source data.

```python
quality_report = []

for name, df in datasets.items():
    quality_report.append({
        "Dataset": name,
        "Rows": df.shape[0],
        "Cols": df.shape[1],
        "Missing values": df.isna().sum().sum(),
        "Missing %": ((df.isna().sum().sum() / (df.shape[0] * df.shape[1])) * 100).round(2),
        "Duplicate Rows": df.duplicated().sum(),
        "Duplicates %": ((df.duplicated().sum() / df.shape[0]) * 100).round(2),
    })

quality_report = pd.DataFrame(quality_report)
quality_report
```

## Initial Findings

The initial assessment indicates that:

- All datasets were successfully loaded.
- Dataset dimensions match the expected structure.
- Some tables contain missing values that require further investigation.
- Duplicate rows are minimal or absent.
- Several columns require datatype conversion.
- The data appears suitable for preprocessing.
